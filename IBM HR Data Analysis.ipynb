# Importing the necessary plotting library for visualization

# count of employees travelling

```python
import seaborn as sns
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='BusinessTravel', palette='Set2')
plt.xlabel('Business Travel')
plt.ylabel('Count')
plt.title('Employee Count by Business Travel')
plt.show()
```

# Department with most Employee's

```python
sns.countplot(data=df,x='Department',palette='husl')
```

# Education field

# Job Role with most Employees

```python
sns.countplot(data=df,x='JobRole',palette='husl')
plt.xticks(rotation=90);
```

# Employees working overtime 

```python
sns.countplot(data=df,x='OverTime',palette='husl')
plt.xticks(rotation=90);
```

# Job Role with most Employees a Bigger Picture

```python
plt.figure(figsize=(25,10))
sns.countplot(data=df, x='JobRole', palette='muted')
plt.xlabel('Job Role')
plt.ylabel('Count')
plt.title('Count of Employees by Job Role')
plt.show()
```

# Donut chart for travel Rate 

```python

labels = df['BusinessTravel'].value_counts().index
sizes = df['BusinessTravel'].value_counts().values


fig, ax = plt.subplots()
ax.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, wedgeprops=dict(width=0.4))


centre_circle = plt.Circle((0, 0), 0.70, fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)


ax.axis('equal')  

plt.title('Business Travel Distribution')
plt.show()
```

# Pie chart for Travel Rate 

```python
travel_counts = df['BusinessTravel'].value_counts()
travel_counts.plot(kind='pie', autopct='%1.1f%%', startangle=200, wedgeprops=dict(width=1), colors=['skyblue', 'lightcoral', 'lightgreen'])
plt.title('Business Travel Distribution')
plt.ylabel('')
plt.show()
```

# Gender wise employment

```python
df['Gender'].value_counts().plot(kind='bar', color=['blue', 'pink'])
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Distribution of Gender')
plt.show()
```

# Employee Count by Gender and Department

```python
df_grouped = df.groupby(['Gender', 'Department']).size().unstack()

df_grouped.plot(kind='bar', stacked=True)
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Employee Count by Gender and Department')
plt.show()
```

# Employee Count by JobRole and Gender

```python
df_grouped = df.groupby(['JobRole', 'Gender']).size().unstack()

df_grouped.plot(kind='bar', stacked=True)
plt.xlabel('JobRole')
plt.ylabel('Employee Count')
plt.title('Employee Count by JobRole and Gender')
plt.show()
```

# Job Satisfaction Rate by Gender

```python
sns.countplot(data=df, x='JobSatisfaction', hue='Gender', palette='husl')
plt.xlabel('Job Satisfaction')
plt.ylabel('Count')
plt.title('Job Satisfaction by Gender')
plt.show()
```
